```python
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
os.listdir('.')
df=pd.read_csv('sub-1003_task-risk_events.csv')
chose_lot=np.logical_and(df.LotisLeft==1,df.response=="['f']")
df["choose_lot"]=chose_lot
display(df.head())
sns.set(rc={'figure.figsize':(15.7,6.27)})

g = sns.pairplot(df, hue="dist")
```

```python
sns.set(rc={'figure.figsize':(15.7,6.27)})

fig=sns.lmplot(x="Lot_per", y="choose_lot", data=df,hue="dist",
           logistic=True, y_jitter=.03);

plt.plot([0, 100], [0.5, 0.5], linewidth=2,linestyle='dashed')

```
